```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
datalist = ['data-1.csv', 'data-2.csv', 'data-3.csv', 'data-4.csv', 'data-5.csv', 'data-6.csv']
descriptionlist = ['fairly strong positive correlation/linear relationship', 
                   'Slightly stronger positive correlation, the data is a little bit more compact.', 
                   '4 Strongly correlated positive linear groups of data. Most likely taken from separate sample groups and combined (stratified random sampling)', 
                   'A border around the outside of the space, with an SFU printed in the bottom right corner.', 
                   '5 groups of vertical lines, with the one the sides more filled in from the center than the ones in the middle.', 
                   'An almost perfectly correlated positive linear relationship with a couple of outliers at point (25,-5). ']
```

```python
def calculate(data):
    xmean = data['x'].mean()
    ymean = data['y'].mean()
    xsd = data['x'].std()
    ysd = data['y'].std()
    xmin = data['x'].min()
    ymin = data['y'].min()
    xmax = data['x'].max()
    ymax = data['y'].max()
    corr = stats.linregress(data['x'], data['y']).rvalue
    return xmean, ymean, xsd, ysd, xmin, ymin, xmax, ymax, corr
```

```python
f = open('summary.txt', 'w')
f.write('')
```

```python
 f = open('summary.txt','a')
```

```python
for i in range(0, 6):
    data = pd.read_csv(datalist[i])
    plt.figure(i)
    plt.scatter(data['x'], data['y'])
    plt.show()
    xmean, ymean, xsd, ysd, xmin, ymin, xmax, ymax, corr = calculate(data)
    f.write('Calculations for data-' + str(i) + ':\n')
    f.write('xmean: ' + str(xmean) + '\n')
    f.write('ymean: ' + str(ymean) + '\n')
    f.write('xsd: ' + str(xsd) + '\n')
    f.write('ysd: ' + str(ysd) + '\n')
    f.write('xmin: ' + str(xmin) + '\n')
    f.write('ymin: ' + str(ymin) + '\n')
    f.write('xmax: ' + str(xmax) + '\n')
    f.write('ymax: ' + str(ymax) + '\n')
    f.write('corr: ' + str(corr) + '\n')
    f.write('description: ' + descriptionlist[i] + '\n\n')
    
```

```python
f.close()
```
